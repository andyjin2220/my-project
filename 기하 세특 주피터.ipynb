```python
import base64
from PIL import Image
from io import BytesIO
import numpy as np
from numpy.linalg import norm
import matplotlib.pyplot as plt
from keras.datasets import mnist

# MNIST 데이터셋 불러오기
(X_train, Y_class_train), (X_test, Y_class_test) = mnist.load_data()

# 유사도 계산 함수 (코사인 유사도)
def cosine_similarity(a, b):
    return np.dot(a.flatten(), b.flatten()) / (norm(a.flatten()) * norm(b.flatten()))

# 테스트셋에서 랜덤으로 10개 이미지 뽑기
#np.random.seed(42)  # 재현성을 위해 시드 설정
random_indices = np.random.choice(X_test.shape[0], 10, replace=False)
random_images = X_test[random_indices]

# 이미지를 이진화하여 저장할 리스트
binarized_test_images = []

for img in random_images:
    # 그레이스케일 변환은 이미 되어 있으므로, 이진화만 수행
    binary_test_image = np.where(img >= 128, 1, 0)
    binarized_test_images.append(binary_test_image)


selected_image_index = np.random.choice(X_train.shape[0])
selected_image = X_train[selected_image_index]

# 선택한 이미지 이진화
binary_selected_image = np.where(selected_image >= 128, 1, 0)

# 유사도 계산
similarities = []
for i, binary_test_image in enumerate(binarized_test_images):
    similarity = cosine_similarity(binary_selected_image, binary_test_image)
    similarities.append(similarity)
    print(f"binary_image {i}: {similarity:.5f}")

# 가장 유사한 이미지를 찾기
most_similar_index = np.argmax(similarities)
most_similar_image = binarized_test_images[most_similar_index]


# 테스트 데이터 이미지들을 수평으로 나열하여 시각화
plt.figure(figsize=(20, 5))
for i, binary_test_image in enumerate(binarized_test_images):
    plt.subplot(2, 10, i+1)
    plt.imshow(binary_test_image, cmap='Greys')
    plt.axis('off')
    plt.title(f"Test {i}\nSim: {similarities[i]:.5f}")

# 학습 데이터와 가장 유사한 테스트 데이터의 이미지를 나열하여 시각화
plt.subplot(2, 10, 11)
plt.imshow(binary_selected_image, cmap='Greys')
plt.axis('off')
plt.title("Selected")

plt.subplot(2, 10, 12)
plt.imshow(most_similar_image, cmap='Greys')
plt.axis('off')
plt.title(f"Most Similar: {similarities[most_similar_index]:.5f}")

plt.tight_layout()
plt.show()
```

유사도 정확도 개선 방법 찾기

2차원 평면 -----> 3차원 도형의 유사도 계산(심화)

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from numpy.linalg import norm

# 격자 크기 설정
grid_size = 30

# 정사면체 내부 확인 함수
def is_point_in_tetrahedron(point, vertices):
    v0, v1, v2, v3 = vertices
    p = point
    def volume(a, b, c, d):
        return np.abs(np.dot(np.cross(b - a, c - a), d - a)) / 6
    v = volume(v0, v1, v2, v3)
    v0p = volume(p, v1, v2, v3)
    v1p = volume(v0, p, v2, v3)
    v2p = volume(v0, v1, p, v3)
    v3p = volume(v0, v1, v2, p)
    return np.isclose(v, v0p + v1p + v2p + v3p)

# 정육면체 내부 확인 함수
def is_point_in_cube(point, center, size):
    x, y, z = point
    cx, cy, cz = center
    half_size = size / 2
    return (cx - half_size <= x <= cx + half_size and
            cy - half_size <= y <= cy + half_size and
            cz - half_size <= z <= cz + half_size)

# 도형을 격자에 추가하는 함수
def add_shape_to_grid(grid, shape_func, **kwargs):
    for x in range(grid_size):
        for y in range(grid_size):
            for z in range(grid_size):
                point = np.array([x / grid_size, y / grid_size, z / grid_size])
                if shape_func(point, **kwargs):
                    grid[x, y, z] = 1
    return grid

# 격자 초기화
cube1_grid = np.zeros((grid_size, grid_size, grid_size), dtype=int)
cube2_grid = np.zeros((grid_size, grid_size, grid_size), dtype=int)
tetrahedron_grid = np.zeros((grid_size, grid_size, grid_size), dtype=int)

# 정육면체1 정의 및 격자에 추가 (중앙에 위치, 부피 0.125)
cube1_center = [0.5, 0.5, 0.5]
cube1_size = 0.5  # 한 변의 길이
cube1_grid = add_shape_to_grid(cube1_grid, is_point_in_cube, center=cube1_center, size=cube1_size)


# 정육면체2 정의 및 격자에 추가 (중앙에 위치, 부피 0.125)
cube2_center = [0.5, 0.5, 0.5]
cube2_size = 0.4  # 한 변의 길이
cube2_grid = add_shape_to_grid(cube2_grid, is_point_in_cube, center=cube2_center, size=cube2_size)


# 정사면체 정의 및 격자에 추가 (중앙에 위치, 부피 0.125에 맞춤)
tetra_height = 0.735  # 정사면체의 높이를 조정하여 부피를 맞춤
tetra_vertices = np.array([
    [0.5, 0.5, 0.5 + tetra_height / 2],
    [0.5 + np.sqrt(3) / 6, 0.5, 0.5 - tetra_height / 2],
    [0.5 - np.sqrt(3) / 12, 0.5 + 0.5, 0.5 - tetra_height / 2],
    [0.5 - np.sqrt(3) / 12, 0.5 - 0.5, 0.5 - tetra_height / 2]
])
tetrahedron_grid = add_shape_to_grid(tetrahedron_grid, is_point_in_tetrahedron, vertices=tetra_vertices)

# 3D 시각화 함수
def plot_voxel_grid(ax, voxel_grid, title):
    ax.voxels(voxel_grid, facecolors='cyan', edgecolor='k')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(title)
    ax.set_xlim(0, grid_size)
    ax.set_ylim(0, grid_size)
    ax.set_zlim(0, grid_size)

fig = plt.figure(figsize=(18, 6))

# 정육면체1 시각화
ax1 = fig.add_subplot(131, projection='3d')
plot_voxel_grid(ax1, cube1_grid, "Voxelized Cube1:0.5")

# 정육면체2 시각화
ax2 = fig.add_subplot(132, projection='3d')
plot_voxel_grid(ax2, cube2_grid, "Voxelized Cube2:0.4")

# 정사면체 시각화
ax3 = fig.add_subplot(133, projection='3d')
plot_voxel_grid(ax3, tetrahedron_grid, "Voxelized Tetrahedron")

plt.tight_layout()
plt.show()

# 3차원 배열을 1차원 벡터로 변환
vector_cube1 = cube1_grid.flatten()
vector_cube2 = cube2_grid.flatten()
vector_tetrahedron = tetrahedron_grid.flatten()

# 코사인 유사도 계산 함수
def cosine_similarity(a, b):
    return np.dot(a, b) / (norm(a) * norm(b))

# 코사인 유사도 계산
similarity = cosine_similarity(vector_cube1, vector_cube2)
print(f"Cosine Similarity(cube1 & cube2): {similarity:.5f}")

# 코사인 유사도 계산
similarity = cosine_similarity(vector_cube1, vector_tetrahedron)
print(f"Cosine Similarity(cube1 & tetrahedron): {similarity:.5f}")

```
